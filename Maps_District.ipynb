```python
# Assign colors based on event type
import math
import folium
import pandas as pd
from IPython.display import display


# Assign colors based on event type
def get_color(event):
    colors = {
        'Diarrhoea': 'red',
        'Influenza-Like Illness': 'blue',
        'Rash': 'orange',
        'Environmental Hazard': 'purple',
        'STI Illness': 'green',
        'Suspected Cholera Case': 'brown',
        #'Suspected Malaria Case': 'lime',
        #'Dog Bite': 'pink',
        #'Ringworms': 'olive',
        #'Suspected TB Case': 'gold',
       # 'Suspected Measles Case': 'crimson',
        'Suspected Case of Mumps': 'lightgreen',
        #'Positive TB Case': 'darkslategrey',
        #'Snake Bite': 'plum',
        #'Unexplained illness': 'khaki',
        #'Malaria Positive Case': 'green',
        'Suspected COVID-19': 'chocolate',
        'Chicken Pox': 'black',
        'Health Education': 'magenta'
    }
    return colors.get(event, 'white')


#Create Map

harare_map = folium.Map(location=[-18.0542, 31.4288], zoom_start=10,tiles = "openstreetmap")

# add marker one by one on the map, and account for Mercator deformation
for index in df.itertuples():
    index_dict = index._asdict()
    local_deformation = math.cos(index.Latitude * math.pi / 180)
    adjusted_radius = (index.affected) * local_deformation
    color = get_color(index.event)
    folium.CircleMarker(
        location=[index.Latitude, index.Longitude],  # Use dot notation to access attributes
        popup='{} ({:.1f})'.format(index.Location,index.affected), # Use dot notation to access attributes
        tooltip = f"Event Recorded: {index.event}, No of Affected: {index.affected}, Location: {index.Location}",
        radius=adjusted_radius,
        color=color,
        fill=True,
        fill_color=color
    ).add_to(harare_map)


# Add title to the map
title_html = '''
            <h3 align="center" style="font-size:20px"><b> Harare Events_June 2024</b></h3>
             '''

# Add title with logos to the map
#title_html = '''
   # <div style="text-align:center;">
     #   <h3 style="font-size:25px;display:inline-block;vertical-align:middle;"><b>Zimbabwe Community Event-Based Surveillance Map</b></h3>
       # <img src="https://i.postimg.cc/sxpkjwjj/GOZ.png" style="height:100px;vertical-align:middle;">
       # <img src="https://i.postimg.cc/Rh7Zk72f/EN-Fundedbythe-EU-RGB-POS.png" style="height:100px;vertical-align:middle;">
       # <img src="https://i.postimg.cc/xjwj94R2/oim-iom-seeklogo-com.png" style="height:100px;vertical-align:middle;">
       # <img src="https://i.postimg.cc/MHnLb9fw/DTM.png" style="height:60px;vertical-align:middle;">

    #</div>#
    #'''

# Add textbox at the bottom right corner
textbox_html = '''
    <div style="position: fixed; bottom: 10px; right: 10px; width: 250px; background-color: white; opacity:0.65; border: 0.5px solid black; padding: 10px; z-index: 1000;">
        <p style="margin: 0;">Source: IOM Zimbabwe | DTM, 2024</p>
        <p style="margin: 0;">Latest data uploaded 10 June 2024.</p>
    </div>
    '''
harare_map.get_root().html.add_child(folium.Element(textbox_html))


harare_map.get_root().html.add_child(folium.Element(title_html))

# Add legend
legend_html = '''
<div style="
    position: fixed;
    bottom: 50px;
    left: 50px;
    width: 250px;
    height: 250px;
    border:2px solid grey;
    z-index:9999;
    font-size:14px;
    background-color:white;
    opacity: 0.85;
    ">
    <b>&nbsp; Legend</b><br><br>
    &nbsp; <i class="fa fa-circle" style="color:red"></i>&nbsp;Diarrhoea<br>
    &nbsp; <i class="fa fa-circle" style="color:blue"></i>&nbsp; Influenza-Type Illness<br>
    &nbsp; <i class="fa fa-circle" style="color:orange"></i>&nbsp; Rash<br>
    &nbsp; <i class="fa fa-circle" style="color:lightgreen"></i>&nbsp; Suspected Case of Mumps<br>
    &nbsp; <i class="fa fa-circle" style="color:purple"></i>&nbsp; Environmental Hazard<br>
    &nbsp; <i class="fa fa-circle" style="color:green"></i>&nbsp; STI Infections<br>
    &nbsp; <i class="fa fa-circle" style="color:chocolate"></i>&nbsp; Suspected COVID-19<br>
    &nbsp; <i class="fa fa-circle" style="color:black"></i>&nbsp; Chicken Pox<br>
    &nbsp; <i class="fa fa-circle" style="color:magenta"></i>&nbsp; Health Education<br>
</div>
'''
harare_map.get_root().html.add_child(folium.Element(legend_html))





# Display the map
display(harare_map)
```
